```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/wfp_food_prices_mwi (3).csv')

df
```

```python
# View the first few rows of the dataset
print(df.head())

# Get a summary of the dataset
print(df.info())

# Get descriptive statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

```

```python
# Reset index to ensure we have a continuous numerical index
df.reset_index(drop=True, inplace=True)

# Remove the first row by index
df = df.drop(index=0)
```

```python
# Filter for entries with "Blantyre" and "maize"
filtered_df = df[(df['district'] == 'Blantyre') & df['commodity'].str.contains('Maize', case=False, na=False)]

# Display the filtered dataset
print(filtered_df.head())
```

```python
# Save the cleaned dataset
filtered_df.to_csv('cleaned_dataset.csv', index=False)
```
